# Working on type B data

```python
# Useful Columns with type B
lst_B = []
temp_B = df[df['type']=='B']
for j in column_list:
    if pd.DataFrame(temp_B[j].unique()).isnull()[0][0]:
        pass
    else:
        lst_B.append(j)
lst_B
df_B = df[df['type']=='B'][lst_B]
df_B
```

```python
plt.figure(figsize=(16,10))
sns.countplot(df_B['batteryCurrent'].sort_values())
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(16,10))
plt.xticks(rotation=90)
sns.countplot(df_B['batteryVoltageAdc'])
```

For frequency chart for type B:
1. Battery current has maximum frequency for 0 Amp
2. Battery Voltage for Analog to Digital convertor lies minimally around 0-30. Maximum frequency can be seen for the battery voltage

```python
plt.figure(figsize=(16,10))
sns.lineplot(df_B['batteryVoltageAdc'],df_B['batteryCurrent'])
```

Minimal correlation between Battery current and Battery Voltage Adc

# Working on type U data

```python
# Useful Columns with type L
lst_L = []
temp_L = df[df['type']=='L']
for j in column_list:
    if pd.DataFrame(temp_L[j].unique()).isnull()[0][0]:
        pass
    else:
        lst_L.append(j)
lst_L
df[df['type']=='L'][lst_L]
```

```python
# Useful Columns with type U
lst_U = []
temp_U = df[df['type']=='U']
for j in column_list:
    if pd.DataFrame(temp_U[j].unique()).isnull()[0][0]:
        pass
    else:
        lst_U.append(j)
lst_U
df[df['type']=='U'][lst_U]
```

```python
df_U = df[df['type']=='U'][lst_U]
```

```python
plt.figure(figsize=(16,10))
sns.lineplot(df_U['throttle'],df_U['batteryCurrent'])
```

From the above graph, an approximately linear relationship can be derived between Battery Current and Throttle percentage

```python
plt.figure(figsize=(16,10))
sns.lineplot(df_U['batteryVoltage'],df_U['batteryCurrent'])
```

 The current only varies when the battery voltage is more than 60

```python
plt.figure(figsize=(16,10))
sns.lineplot(df_U['wheelRpm'],df_U['batteryCurrent'])
```

1. The battery current increases linearly with Wheel RPM until 600 RPM
2. The battery current decreases linearly with Wheel RPM from 600 RPM to 900 RPM
3. After 900 RPM it is difficult to derive a relationship between Wheel RPM and Battery Current

```python
plt.figure(figsize=(16,10))
sns.lineplot(df_U['controllerTemperature'],df_U['batteryCurrent'])
```

A correlation can be observed. When controller Temperature increases above 45 units, the battery current is seen to be rising. Conversely, a increasing controller temperature can imply a higher current being derived from the battery.

```python
sns.heatmap(df[['throttle','batteryCurrent','batteryVoltage','controllerTemperature']].corr())
```

1. A higher correlation can be observed between Battery current and throttle
2. A correlation can also be observed between throttle and controller temperature
3. A correlation can also be observed between Battery current and controller temperature
4. Battery Voltage has a very poor correlation with any of the continuous values
